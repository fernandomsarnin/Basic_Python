```python
import matplotlib.pyplot as plt
from cycler import cycler

cores= plt.get_cmap('Set3').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)


plt.figure(figsize=(8, 5))

produtos = ['TV','Celular','Tablet','Mouse','Teclado','Notebook', 'Monitor', 'Ipad']
estoque = [1500, 1680, 800, 3200, 758, 2054, 985, 580]

nivel_min_estoque = 1000

lista_produto = []
lista_estoque = []

for i, produto in enumerate(produtos):
    if estoque[i] < nivel_min_estoque:        
        print('O produto {} está com apenas {} unidades em estoque. Informar com urgência o departamento de compras!'.format(produto, estoque[i]))
        lista_produto.append(produto)
        lista_estoque.append(estoque[i])
        barras = plt.bar(produto, estoque[i])              
        plt.bar_label(barras, font='Calibri',fontsize=12, fontweight='bold', color='white', padding=-15)
        plt.tick_params(axis='x', length=0)
        plt.xticks(font='Calibri', fontsize=12, rotation=20)
        plt.yticks([])
        plt.title('Produtos com Quantidade Abaixo do Nível Mínimo de Estoque!', font='Calibri', fontsize=15, fontweight='bold')
        plt.box(False)
        plt.show
```

```python
plt.figure(figsize=(8, 5))
barras = plt.bar(lista_produto, lista_estoque, color='gray', alpha=0.5)
barras[3].set_color('darkred')
barras[3].set_alpha(0.8)
plt.box(False)
plt.yticks([])
plt.tick_params(axis='x', length=0, rotation=20)
plt.title('Produtos com Quantidade Abaixo do Nível Mínimo de Estoque!', font='Calibri', fontsize=15, fontweight='bold')
plt.bar_label(barras, font='Calibri',fontsize=12, fontweight='bold', color='white', padding=-15)
plt.show()
```

```python
destaques = [0,0,0,0.1]

plt.figure(figsize=(12,6))
plt.title('Produtos com Quantidade Abaixo do Nível Mínimo de Estoque!', font='Calibri', fontsize=15, fontweight='bold')
plt.pie(lista_estoque, labels=lista_produto, autopct='%.1f', explode=destaques, shadow=True)


plt.show
```
